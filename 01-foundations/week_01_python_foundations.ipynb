```python
"""
🌸 Project: Seoul Startup 2027 Path
📅 Day 1: The Founder Intake Logic
---------------------------------------------------------
Today, I officially started my 365-day streak for the
K-Startup Grand Challenge 2027.

What I mastered and applied today:
✅ Function Design: Using 'def main()' to organize my logic.
✅ Data Sanitization: Cleaning user input with .strip() and .capitalize().
✅ Type Casting: Converting 'str' to 'int' to perform logic checks.
✅ Error Resilience: Using 'try/except' so my program doesn't crash.
✅ The Ignition Switch: Using 'if __name__ == "__main__"' like a pro.
---------------------------------------------------------
"""

def main():
    # Let's get the name and make it look professional (No messy spaces!)
    name = input("Enter your Name: ").strip().capitalize()

    try:
        # We need a number for age, so we convert (str to int)
        age = int(input("Enter your Age: "))

        # Scenario A: The user made a mistake and typed 0 or negative
        if age <= 0:
            print(f"Hey {name}, Please enter a valid age! Try again.")

        # Scenario B: Future founders (Under 18)
        elif age < 18:
            print(f"Hello {name}! You've got the spark, but you'll need a Mentor 🗣️")

        # Scenario C: The 'Sweet Spot' for the K-Startup Grand Challenge
        elif 18 <= age <= 30:
            print(f"Perfect timing, {name}! You are in the 'Prime Window' for Seoul 🚀")

        # Scenario D: The experienced founders
        else:
            print(f"Welcome, {name}. Your experience is a massive asset for 2027 🎯")

    except ValueError:
        # This catches the 'Oops' if the user types "Twenty" instead of 20
        print("Wait! Please enter your age using numbers only.")

# This is the "Ignition Switch" that starts the program
if __name__ == "__main__":
    main()
```

```python
"""
📅 Day 2: The Batch Processing Engine
---------------------------------------------------------
Today, I moved from handling one user to building a scalable
system that can process multiple applicants at once.

What I mastered and applied today:
✅ Data Structures: Using 'Dictionaries' as a mini-database.
✅ Iteration: Using 'for loops' to automate repetitive tasks.
✅ Key-Value Mapping: Storing names as 'Keys' and ages as 'Values'.
✅ Batch Logic: Processing a full list of applicants in one run.
✅ Founder Detection: Adding custom conditional logic for system admin.
---------------------------------------------------------
"""
def startup_report():
    # Step 1: Create the Dictionary (database)
    applicants = {}

    # Step 2: Data Collection Phase
    print("📋 Data Entry Phase (Please enter 3 applicants)")
    for i in range(3):
        name = input(f"Enter name for Applicant {i+1}: ").strip().title()
        age = int(input(f"Enter age for {name}: "))
        applicants[name] = age # Saving to our dictionary

    # Step 3: The Automated Report Phase
    print("\n 📊 STARTUP ELIGIBILITY REPORT \n")

    for name, age in applicants.items():
        if name == "Anoosha":
            print(f"🌟 FOUNDER: {name} | Status: Access Granted")
        elif age >= 18:
            print(f"✅ Applicant: {name} | Age: {age} | Status: Eligible")
        else:
            print(f"❌ Applicant: {name} | Age: {age} | Status: Underage")

    print("End of Report.")

if __name__ == "__main__":
    startup_report()
```

```python
"""
📅 Day 3: Mastering Functional Architecture & Delegation
---------------------------------------------------------
Today's breakthrough: Moving from 'Linear Coding' to 'Modular Systems'.
By separating logic into specialized functions, I've made Project Alpha
scalable and easier to secure.

Key implementations:
✅ Multi-Parameter Functions (Handling Price + Tax)
✅ Data Flow (Passing values from the 'CEO' main to 'Employee' functions)
---------------------------------------------------------
"""


# CEO: ORDER OF OPERATIONS
def main():
    print("🚀 --- PROJECT ALPHA SYSTEM CHECK --- 🚀\n")

    # 1. Multiplier Drill (Testing Basic Functions)
    num1 = int(input("Enter a number to multiply by 10: "))
    results = multiple(num1)
    print(f"Result: {num1} × 10 = {results}\n")

    # 2. Currency Drill (Bridge to Korea)
    pkr = int(input("Enter PKR value to convert to Won: "))
    won_result = to_won(pkr)
    print(f"Result: {pkr} PKR in Won is {won_result}\n")

    # 3. Security Drill (Password Gate)
    password = input("Enter a 8-digit password to check: ")
    security = pass_check(password)
    print(f"Result: Your password is {security}\n")

    # 4. Math Drill (Book Price + Tax)
    price = int(input("Enter the book price: "))
    tax = int(input("Enter the tax amount: "))
    final_price = total(price, tax)
    print(f"Result: Total price of the book is: {final_price}\n")

    print("✅ --- ALL DRILLS COMPLETED SUCCESSFULLY ---")


# EMPLOYEES: DO THE WORK

def multiple(num):
    # Just multiplies by 10
    return num * 10

def to_won(n):
    # Converts PKR to Korean Won
    return n * 4.5


def pass_check(n):
    # Security gate: Checks if password is at least 8 digits
    if len(n) >= 8:
        return "Strong"
    else:
        return "Weak"

def total(p, t):
    # Adds price and tax together
    return p + t


# THE IGNITION SWITCH
if __name__ == "__main__":
    main()
```

```python
"""
📅 Day 4: The Resilient Percent Calculator
---------------------------------------------------------
Focus: Exception Handling (try/except) & Input Validation
Goal: Prevent system crashes when students enter non-numeric data.

This module ensures that Project Alpha's data entry remains
100% stable even if a user makes a typing error.
---------------------------------------------------------
"""

# THE CEO (Main Logic with Error Handling)
def main():
    print("🛡️ PROJECT ALPHA: DATA INTEGRITY GATEWAY 🛡️\n")

    # Use a loop to force a valid entry without crashing
    while True:
        try:
            # We take the input and try to convert it immediately
            user_input = input("Enter your score out of 100: ")
            user_score = float(user_input)

            # If float() succeeds, we break the loop
            break
        except ValueError:
            # If it fails, we use 'user_input' because 'user_score' wasn't created
            print(f"❌ Error: '{user_input}' is not a valid number. Please try again.\n")

    # Calculate and display the result only after valid input is confirmed
    result = percent(user_score)
    print(f"✅ Success! Your calculated percentage is: {result}%")


# THE EMPLOYEES (The Logic)

def percent(n):
    """Calculates the percentage based on a total of 100."""
    return (n / 100) * 100


# START THE ENGINE
if __name__ == "__main__":
    main()
```

```python
"""
Day 5: The Persistent Data Vault
---------------------------------------------------------
Today, I moved from "temporary memory" to a "permanent vault."
I built a system that saves data into a file so it's still
there even after I turn off my computer.

What I mastered today:
✅ Keeping data forever using the 'csv' module.
✅ Making sure files close safely with 'with'.
✅ Setting up "Headers" (Labels) so the data stays organized.
✅ Adding new info without deleting the old stuff.
---------------------------------------------------------
"""

import csv

# Creating the permanent file
# "w" means we are creating a fresh file with our top labels

with open("tutoring_log.csv", "w", newline = "") as file:
    # This 'writer' is like a specialized pen for CSV files
    writer = csv.writer(file)

    # 1. We write the labels once at the top so the columns have names
    writer.writerow(["Student Name", "Class", "Grade", "Hours"])

    # 2. We use a loop to ask for 2 students in a row
    for i in range(2):
        print(f"\n--- Enter Student {i+1} Details ---")
        name = input("Name: ")
        class_ = int(input("Class: "))
        grade = input("Grades: ")
        hours = float(input("Hours: "))

        # This saves the student's info as a new row in our file
        writer.writerow([name, class_, grade, hours])

# Final confirmation that our data is safe!
print("\nFile 📁 created successfully ✅")
```

```python
"""
Day 6: Unit Testing & Libraries
---------------------------------------------------------
Today, I upgraded my system from "Storage" to "Analysis."
I learned how to use professional libraries to calculate
data and how to use 'assertions' to verify my code's accuracy.

What I mastered today:
✅ Library Imports: Using 'statistics' for professional math.
✅ Error Handling: Using 'try/except' to prevent app crashes.
✅ Unit Testing: Using 'assert' to prove my logic works.
✅ Clean Architecture: Separating 'Input' logic from 'Calculation' logic.
---------------------------------------------------------
"""

from statistics import mean, median

def main():
    # Step 1: Ask for Input
    user1 = input("Enter your first score: ")
    user2 = input("Enter your second score: ")

    try:
        # Step 2: Convert text to numbers safely
        score1 = float(user1)
        score2 = float(user2)

        data = [score1, score2]

        # Step 3: Get results from our function
        mean_value, median_value = result(data)

        print(f"The mean of the scores is: {mean_value}")
        print(f"The median of the scores is: {median_value}")

    except ValueError:
        print("❌ Invalid input! Please enter numbers for the scores.")

def result(scores):
    # Does the Math
    return mean(scores), median(scores)

# Unit Testing
# We test our engine with known values to make sure it's 100% correct
assert result([10, 20]) == (15.0, 15.0)
assert result([0, 100]) == (50.0, 50.0)
print("The Test is PASSED ✅. Architecture is solid.")

# THE IGNITION SWITCH
if __name__ == "__main__":
    main()
```

```python
"""
Day 7: Regular Expression (RegEx)
---------------------------------------------------------
Today, I tackled the "alien language" of Regular Expressions.
I moved from searching for specific words to searching for "patterns" within text.

What I mastered today:
✅ The 're' Library: Python's built-in tool for pattern matching.
✅ Anchors: Using '^' and '$' to lock the start and end of a string.
✅ Quantifiers: Understanding how '.', '*', and '+' change a search.
---------------------------------------------------------
"""

import re

def validate_system():
    # The Pattern: 2 Letters + 3 Numbers (e.g. AK123)
    pattern = r"^[A-Z]{2}[0-9]{3}$"

    print("📋 Student ID Validation...")
    while True:
        student_id = input("Enter Student ID: ").strip()

    # Checking if the ID matches the required architecture
        if re.match(pattern, student_id):
            print(f"✅ VALID: {student_id} matches the system schema. Access Granted.")
            break
        else:
            print("❌ INVALID: Format must be 2 Capital Letters followed by 3 Digits.")

if __name__ == "__main__":
    validate_system()
```
