```python
"""
🌸 Project: Seoul Startup 2027 Path
📅 Day 1: The Founder Intake Logic
---------------------------------------------------------
Today, I officially started my 365-day streak for the
K-Startup Grand Challenge 2027.

What I mastered and applied today:
✅ Function Design: Using 'def main()' to organize my logic.
✅ Data Sanitization: Cleaning user input with .strip() and .capitalize().
✅ Type Casting: Converting 'str' to 'int' to perform logic checks.
✅ Error Resilience: Using 'try/except' so my program doesn't crash.
✅ The Ignition Switch: Using 'if __name__ == "__main__"' like a pro.
---------------------------------------------------------
"""

def main():
    # Let's get the name and make it look professional (No messy spaces!)
    name = input("Enter your Name: ").strip().capitalize()

    try:
        # We need a number for age, so we convert (str to int)
        age = int(input("Enter your Age: "))

        # Scenario A: The user made a mistake and typed 0 or negative
        if age <= 0:
            print(f"Hey {name}, Please enter a valid age! Try again.")

        # Scenario B: Future founders (Under 18)
        elif age < 18:
            print(f"Hello {name}! You've got the spark, but you'll need a Mentor 🗣️")

        # Scenario C: The 'Sweet Spot' for the K-Startup Grand Challenge
        elif 18 <= age <= 30:
            print(f"Perfect timing, {name}! You are in the 'Prime Window' for Seoul 🚀")

        # Scenario D: The experienced founders
        else:
            print(f"Welcome, {name}. Your experience is a massive asset for 2027 🎯")

    except ValueError:
        # This catches the 'Oops' if the user types "Twenty" instead of 20
        print("Wait! Please enter your age using numbers only.")

# This is the "Ignition Switch" that starts the program
if __name__ == "__main__":
    main()
```

```python
"""
📅 Day 2: The Batch Processing Engine
---------------------------------------------------------
Today, I moved from handling one user to building a scalable
system that can process multiple applicants at once.

What I mastered and applied today:
✅ Data Structures: Using 'Dictionaries' as a mini-database.
✅ Iteration: Using 'for loops' to automate repetitive tasks.
✅ Key-Value Mapping: Storing names as 'Keys' and ages as 'Values'.
✅ Batch Logic: Processing a full list of applicants in one run.
✅ Founder Detection: Adding custom conditional logic for system admin.
---------------------------------------------------------
"""
def startup_report():
    # Step 1: Create the Dictionary (database)
    applicants = {}

    # Step 2: Data Collection Phase
    print("📋 Data Entry Phase (Please enter 3 applicants)")
    for i in range(3):
        name = input(f"Enter name for Applicant {i+1}: ").strip().title()
        age = int(input(f"Enter age for {name}: "))
        applicants[name] = age # Saving to our dictionary

    # Step 3: The Automated Report Phase
    print("\n 📊 STARTUP ELIGIBILITY REPORT \n")

    for name, age in applicants.items():
        if name == "Anoosha":
            print(f"🌟 FOUNDER: {name} | Status: Access Granted")
        elif age >= 18:
            print(f"✅ Applicant: {name} | Age: {age} | Status: Eligible")
        else:
            print(f"❌ Applicant: {name} | Age: {age} | Status: Underage")

    print("End of Report.")

if __name__ == "__main__":
    startup_report()
```

```python
"""
📅 Day 3: Mastering Functional Architecture & Delegation
---------------------------------------------------------
Today's breakthrough: Moving from 'Linear Coding' to 'Modular Systems'.
By separating logic into specialized functions, I've made Project Alpha
scalable and easier to secure.

Key implementations:
✅ Multi-Parameter Functions (Handling Price + Tax)
✅ Data Flow (Passing values from the 'CEO' main to 'Employee' functions)
---------------------------------------------------------
"""


# CEO: ORDER OF OPERATIONS
def main():
    print("🚀 --- PROJECT ALPHA SYSTEM CHECK --- 🚀\n")

    # 1. Multiplier Drill (Testing Basic Functions)
    num1 = int(input("Enter a number to multiply by 10: "))
    results = multiple(num1)
    print(f"Result: {num1} × 10 = {results}\n")

    # 2. Currency Drill (Bridge to Korea)
    pkr = int(input("Enter PKR value to convert to Won: "))
    won_result = to_won(pkr)
    print(f"Result: {pkr} PKR in Won is {won_result}\n")

    # 3. Security Drill (Password Gate)
    password = input("Enter a 8-digit password to check: ")
    security = pass_check(password)
    print(f"Result: Your password is {security}\n")

    # 4. Math Drill (Book Price + Tax)
    price = int(input("Enter the book price: "))
    tax = int(input("Enter the tax amount: "))
    final_price = total(price, tax)
    print(f"Result: Total price of the book is: {final_price}\n")

    print("✅ --- ALL DRILLS COMPLETED SUCCESSFULLY ---")


# EMPLOYEES: DO THE WORK

def multiple(num):
    # Just multiplies by 10
    return num * 10

def to_won(n):
    # Converts PKR to Korean Won
    return n * 4.5


def pass_check(n):
    # Security gate: Checks if password is at least 8 digits
    if len(n) >= 8:
        return "Strong"
    else:
        return "Weak"

def total(p, t):
    # Adds price and tax together
    return p + t


# THE IGNITION SWITCH
main()
```
