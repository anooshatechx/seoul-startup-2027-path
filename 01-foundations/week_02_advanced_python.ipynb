```python
"""
Day 8: Logic Persistence & Pattern Architecture
---------------------------------------------------------
Today, I moved from simple validation to a persistent qschema.
I implemented a multi-stage input loop that ensures every piece of data
is perfect before it enters the application.

What I mastered today:
✅ Persistence Loops: Using 'while True' and 'break' to force valid inputs.
✅ Optional Quantifiers: Using '?' to handle flexible data patterns.
✅ Modular Validation: Creating specialized functions for distinct data.
✅ Case Insensitivity: Using 're.IGNORECASE' for flexibility.
---------------------------------------------------------
"""

import re

def main():
    print(f"------- STUDENT RECORD CHECKING -------\n")

    # Persistent Validation for ID
    while True:
        user_id = input("Enter StudentID (i.e. AB123): ").strip()
        if valid_id(user_id):
            break

    # Persistent Validation for Subject Code (Optional Numbers)
    while True:
        subject = input("Enter SubjectCode (i.e. #MTH201): ").strip()
        if subject_code(subject):
            break

    # Persistent Validation for Status
    while True:
        status = input("Status (Active/Graduated): ").strip()
        if current_status(status):
            break

    print("\n------- STUDENT RECORD CHECKED ✅ -------")

def valid_id(i):
    # Checks for 2 Uppercase letters followed by 3 numbers.
    if match := re.search(r"^[A-Z]{2}[0-9]{3}$", i):
        print("✅ ID successful")
        return match
    else:
        print("❌ Invalid ID...")

def subject_code(j):
    # Checks for # + 3-4 uppercase letters + optional 3-4 numbers.
    if match := re.search(r"^#[A-Z]{3,4}([0-9]{3,4})?$", j):
        print("✅ SubjectCode successful")
        return match
    else:
        print("❌ Invalid SubjectCode...")

def current_status(k):
    # Checks for specific status regardless of capitalization.
    if match := re.search(r"^Active|Graduated$", k, re.IGNORECASE):
        print("✅ Status successful")
        return match
    else:
        print("❌ Invalid Status...")


if __name__ == "__main__":
    main()
```

```python
"""
Day 9: Object-Oriented Architecture
---------------------------------------------------------
Today, I transitioned from flat logic to a scalable Object-Oriented (OOP)
architecture. I implemented a 'Concept' Class that acts as the blueprint
for every lesson in Project Alpha.

What I mastered today:
✅ Class Construction: Defining the 'Table Schema' for academic concepts.
✅ Attributes (Columns): Mapping English terms to Native Bridges.
✅ State Management: Adding 'is_mastered' to track student progress locally.
---------------------------------------------------------
"""

class Concept:
    # The blueprint for Project Alpha lessons.
    def __init__(self, english_word, native_bridge, visual_data):

        # Attribute Initialization (Columns)
        self.english_word = english_word
        self.native_bridge = native_bridge
        self.visual_data = visual_data
    def display_lesson(self):
        # Displays the lesson in a human-centric format.
        print(f"English Term: {self.english_word}")
        print(f"Native Bridge: {self.native_bridge}")
        print(f"Visual Logic: {self.visual_data}")
        print(f"---------------------------------------")

def main():
    # Creating the first 'Row' (Object)

    # Lesson 1: Fetch
    print(f"\n------- 🛡️ PROJECT ALPHA LESSON -------\n")
    lesson_1 = Concept(
        english_word="Fetch",
        native_bridge="Laana (لانا)",
        visual_data="Someone bringing a book to you 📕"
    )

    # Lesson 2: Clumsy
    lesson_2 = Concept(
        english_word="Clumsy",
        native_bridge="Sharminda hona (Situational: dropping things)",
        visual_data="Dropping books while others watch"
)

    # Executing the Lesson Bridge
    lesson_1.display_lesson()
    lesson_2.display_lesson()

if __name__ == "__main__":
    main()
```

```python
"""
Day 10: State Management & Logic
---------------------------------------------------------
Today I worked on making my classes smarter. Instead of doing the math
outside the class, I put the logic right inside the object. It makes
the code much cleaner and easier to manage.

What I mastered today:
✅ Encapsulation: Keeping the math inside the class where it belongs.
✅ String Formatting: Using __str__ so I can just print the object directly.
✅ User Input: Connecting the terminal to my object attributes.
---------------------------------------------------------
"""

class Student:
    def __init__(self, name, current_subject, progress):
        self.name = name
        self.current_subject = current_subject
        self.progress = progress

    def progress_left(self):
        # Calculate how much is left to finish
        return 100 - self.progress

    def __str__(self):
        # This makes printing the object look nice and readable
        return f"Student {self.name} is studying {self.current_subject} and needs {self.progress_left()}% more to finish!"

def main():
    name = input("Enter name: ")
    current_subject = input("Enter Subject: ")
    progress = int(input("Enter progress value: "))

    my_student = Student(name, current_subject, progress)
    print(my_student)

if __name__ == "__main__":
    main()
```

```python
"""
Day 11: Pattern Matching & Active Learning
---------------------------------------------------------
Today I combined classes with regex to build a smart quiz. It’s cool
because the code now checks my SQL queries for me. If I get it right,
the object updates itself automatically.

What I mastered today:
✅ Regex: Used it to catch different ways to type a command.
✅ Class Logic: Encapsulated the mastery check inside the class.
✅ Feedback Loop: Used a while loop to keep me practicing until I win.
---------------------------------------------------------
"""

import re

class Concept:
    def __init__(self, title, category, mastery, amount):
        self.title = title
        self.category = category
        self.mastery = mastery
        self.amount = amount

    def is_learned(self):
        # Returns 1 if mastery is 8 or higher, else 0
        return 1 if self.mastery >= 8 else 0

    def improve_mastery(self, amount):
        # Boosts the mastery score and caps it at 10
        self.mastery += self.amount
        if self.mastery > 10:
            self.mastery = 10

    def __str__(self):
        # Shows current status in a clean format
        return f"Topic: {self.title} | Mastery: {self.mastery}/10 | Learned: {self.is_learned()}"

def main():
    title = input("Enter topic: ")
    category = input("Which category?: ")
    mastery = int(input("Level of mastery (1-10): "))

    # Initialize the object
    tracker = Concept(title, category, mastery, 1)

    while True:
        # We check the object's current mastery level in the loop
        if tracker.mastery < 8:
            question = input("What will be the query for selecting all the columns of table Teacher?: ")

            # Check the input against the required SQL pattern
            if re.match(r'^(Select|SELECT|select) \* (FROM|from|From) (Teacher|TEACHER|teacher)$', question):
                tracker.improve_mastery()
                print("Correct! You got this 🏆")
                break
            else:
                print("Think Harder!")
        else:
            print("Great! You're already good at it 🏆")
            break

    # Print the final status
    print(tracker)

if __name__ == "__main__":
    main()
```
