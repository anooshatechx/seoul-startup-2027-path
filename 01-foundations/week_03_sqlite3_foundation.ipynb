```python
# Import SQLite3 
import sqlite3

# Create Class
class Practice():
    def __init__(self, id, project_name, language):
        self.id = id
        self.project_name = project_name
        self.language = language

# Connect to Database
conn = sqlite3.connect("simple_practice.db")
cur = conn.cursor()

# Create Table
cur.execute("""CREATE TABLE IF NOT EXISTS Practice_Database(
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Project_Name TEXT,
    Language TEXT)""")

# Create Function for Values
def add_values(project_name, language):
    cur.execute("INSERT INTO Practice_Database VALUES (?,?,?)", (None,project_name,language))
    id = cur.lastrowid
    conn.commit()

    print(f"🚀 Successfully inserted '{project_name}' into the database with ID: {id}")

# Insert Values
add_values("Kognit","Python")

conn.close()

print("🔒 Database connection closed safely. All changes saved!")
```
