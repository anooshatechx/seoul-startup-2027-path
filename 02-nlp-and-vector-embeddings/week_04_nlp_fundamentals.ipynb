```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I love hugging face.")



### 3. Transformer Models & Architectures
Transformers are language models trained on massive datasets. They have three architectural categories:

#### ① Encoder-Only Models
* **Directionality:** Bi-directional (understands context from both the left and right sides of a word simultaneously).
* **Core Strength:** Excellent at understanding text and comprehension tasks.
* **Training Type:** Masked Language Modeling (MLM).
* **Examples:** BERT, RoBERTs.

#### ② Decoder-Only Models
* **Directionality:** Uni-directional (predicts the next word based only on previous word).
* **Core Strength:** Excellent for text generation tasks.
* **Training Type:** Causal Language Modeling (CLM).
* **Examples:** GPT series, LLaMA.

#### ③ Encoder-Decoder Models (Sequence-to-Sequence)
* **Mechanism:** The input is processed by the encoder, and the final output is generated by the decoder.
* **Core Strength:** Perfect for transforming text from one form to another.
* **Tasks:** Translation, summarization.
* **Examples:** T5, BART.

---

### 4. End-to-End Pipeline: How Message Becomes Model Output
When a user interacts with a Transformer model, the input text goes through a multi-step pipeline before producing a final output result:

```text
[ User Text ] ➔ [ Tokenization ] ➔ [ Embedding ] ➔ [ Transformer Layers ] ➔ [ Output Generation]

* **User Sends Text:** For example: `"I love hugging face!"`. Computers cannot read raw text directly, they understand in the form of vectors.
* **Tokenization:** The raw text is split into smaller units called tokens. Each individual token is given a unique ID number.
  * *Example:* `["I", "love", "Hugging", "Face", "!"]` ➔ `[101, 1045, 2293, 17662, 999, 102]`
* **Embedding:** These token IDs are then converted into vectors (lists of numbers). This step ensures words with similar meanings are kept closer together, allowing the model to grasp underlying relationships.
* **Transformer Layers:** The encoder and/or decoder layers apply **attention mechanisms**. This step allows the model to learn context and determine exactly "who relates to what" across the entire sentence.
* **Output Generation:** The model processes the contextual representations to predict and deliver the final output result.

---

### 5. Hugging Face `pipeline()` Task Capabilities
The `pipeline` abstraction in Hugging Face supports a diverse set of built-in NLP tasks:

* **1. sentiment-analysis:** Classifies text into positive (`+ve`) or negative (`-ve`) sentiment.
* **2. zero-shot-classification:** Allows you to select labels for classification.
* **3. text-generation:** Uses an input prompt to generate text.
* **4. fill-mask:** Predicts and fills in a missing or masked word inside a sentence.
* **5. ner (Named Entity Recognition):** Identifies real-world entities like a `Person`, `Location (loc)`, or `Organization (org)` inside a sentence.
* **6. question-answering:** Extracts accurate answers from a provided question text.
* **7. summarization:** Compresses long-form documents into concise summaries.
* **8. translation:** Translates text from one language to another.
