```python
# Import the required libraries
from sentence_transformers import SentenceTransformer,util

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode the question and answers
question = model.encode("What is scarcity")
answers = model.encode(["You have 1 chocolate and you have to give it to 14 hungry students in your class so the item (chocolate) is less than the people (student).", "I have 10 apples"])

# Perform semantic search to find the most relevant answer
print(util.semantic_search(question, answers))
```
