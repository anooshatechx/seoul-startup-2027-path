# Kognit Vector Search Flow

## Step 1: Install Libraries

We need three libraries:

- **sqlite3:** because we need the analogies from our database
- **sqlite_vec:** because we have to create a virtual table for vector storage
- **sentence_transformer:** to convert words into numbers (vectors)

## Step 2: Connect Database to sqlite-vec

Connect our sqlite3 database and give it to sqlite_vec so it gains vector search powers.

## Step 3: Create Virtual Table

Create a virtual table so we can store numbers (vectors) easily and search them by meaning.

## Step 4: Encode and Store Analogies

Convert words into numbers by embedding using sentence_transformer model encoding.

- Take colloquial English text from our analogies table
- Encode it into 384 numbers (the vector)
- Send those vectors to the virtual table for storage
